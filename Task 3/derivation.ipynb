
$\textbf{Derivation of Moment of Inertia for a Solid Rectangle:}$

$\text{Consider a differential strip of thickness} \ dy \ \text{at a distance} \ y \ \text{from the base of the rectangle.} $

$\text{The area of this strip,} \ dA, \text{is} \ b \times dy.$

$\text{The differential moment of inertia of this strip about the centroidal axis is:}$

$dI = b \times y^2 \times dy$

$\text{To find the total moment of inertia of the rectangle, integrate} \ dI \ \text{over the height of the rectangle:} $

$I = \int_0^h b \times y^2 \times dy $

$I = b \int_0^h y^2 \, dy $

$I = b \left[ \frac{y^3}{3} \right]_0^h $

$I = b \left[ \frac{h^3}{3} - 0 \right] $

$I = \frac{b \times h^3}{3} $ 

$\text{However, this is the moment of inertia about the base of the rectangle. The parallel axis theorem states:} $

$ I_{\text{centroid}} = I_{\text{base}} - A \times \left( \frac{h}{2} \right)^2 $

$\text{Where} \ A \ \text{is the area of the rectangle,} \ b \times h $

$\text{Substituting in the values:} $

$I_{\text{centroid}} = \frac{b \times h^3}{3} - b \times h \times \left( \frac{h}{2} \right)^2 $

$I_{\text{centroid}} = \frac{b \times h^3}{3} - \frac{b \times h^3}{4} $

$I_{\text{centroid}} = \frac{b \times h^3}{12}$

$\text{So, the moment of inertia of the rectangle about its centroid is:} $

$I = \frac{b \times h^3}{12}$



The area of the cross section is: 

$A = b \times b = 100mm \times 100mm = 10000 mm^2$

The moment of inertia is:

$I = \frac{b \times h^3}{12} = \frac{100 \times 100^3}{12} = 8333333.33 mm^4 = 8.33333333 \times 10^{-6} m^4 $

The shape factor is:

$\phi^e_B = \frac{12I}{A^2} = \frac{12 \times 8333333.33}{10000^2} = 0.9999999996 = 1$

This is because the shape is a square cross-section area so it is the same as the referenced shape.

The Young's modulus of glass-fibre reinforced bar (GRP) is:

$E = 2.4 GPa = 2.4e9 Pa$. Let $C = \frac{1}{L^3}$

The effective stiffness of simply supported beam is 

$ S_0 =\frac{3EI}{L^3} = \frac{3 \times 2.4e9 Pa \times 8.33333333 \times 10^{-6} m^4}{L^3} = C \times 60000 N/m$

The shape factor is 1, so the bending stiffness is the same as the referenced shape: $S = S_0$

$\textbf{Derivation of Moment of Inertia for a Hollow rectangle Section:}$

$\text{Given a hollow square with outer width} \ b \ \text{and height} \ h, \text{and thickness} \ t$

$\text{the inner square has width} \ b - 2t \ \text{and height} \ h - 2t. $

$\text{For the outer square:} $

$I_{\text{outer}} = \frac{b \times h^3}{12}$

$\text{For the inner square:} $

$I_{\text{inner}} = \frac{(b - 2t) \times (h - 2t)^3}{12}$

$\text{The moment of inertia of the hollow section is:} $

$I_{\text{hollow}} = I_{\text{outer}} - I_{\text{inner}}$

$\text{Substituting in the values and simplifying, we get:} $

$I_{\text{hollow}} = \frac{1}{6} h^3 t (1 + 3\frac{b}{h})$


The area of the cross section is: 

$A = 2 \times t (b + h - 2t) = 2 \times 10mm (100mm + 100mm - 2 \times 10mm) = 3600 mm^2$

The moment of inertia is:

$I = \frac{1}{6} h^3 t (1 + 3\frac{b}{h}) =  \frac{1}{6} (100mm)^3 (10mm) (1 + 3\frac{100mm}{100mm}) = 6666666.66 mm^4 = 6666666.66 \times 10^{-6} m^4 $

The shape factor is:

$\phi^e_B = \frac{12I}{A^2} = \frac{12 \times 6666666.66}{3600^2} = 6.172$

The effective stiffness of simply supported beam is 

$ S_0 =\frac{3EI}{L^3} = \frac{3 \times 2.4e9 Pa \times 8.33333333 \times 10^{-6} m^4}{L^3} = C \times 60000 N/m$

The shape factor is $6.172$, so the bending stiffness of this shape is

$ S = 6.172 \times S_0 = C \times 370370 N/m$

$\textbf{Derivation of Moment of Inertia for a Hollow Circular Cross-Section:}$

$\text{Given a hollow circle with outer radius} \ r_o \ \text{and inner radius} \ r_i, \text{and thickness} \ t = r_o - r_i.$

$\text{The moment of inertia} \ I \ \text{of a circle about an axis perpendicular to its plane is:}$ 

$I = \frac{\pi \times r^4}{4}$

$\text{Using this formula:}$

$\text{For the outer circle:} $

$I_{\text{outer}} = \frac{\pi \times r_o^4}{4}$

$\text{For the inner circle:}$

$I_{\text{inner}} = \frac{\pi \times r_i^4}{4}$

$\text{The moment of inertia of the hollow section is:} $

$I_{\text{hollow}} = I_{\text{outer}} - I_{\text{inner}}$

$\text{Substituting in the values and simplifying, we get:}$

$I_{\text{hollow}} = \frac{\pi}{4} (r_o^4 - r_i^4)$


The area of the cross section is: 

$A = \pi (r_o^2 - r_i^2) = \pi ((250mm)^2 - (240mm)^2) = 15393.8 mm^2$

The moment of inertia is:

$I = \frac{\pi}{4} (r_o^4 - r_i^4) = \frac{\pi}{4} ((250mm)^4 - (240mm)^4) = 462198965.178 mm^4 = 4.622 \times 10^{-4} m^4 $

The shape factor is:

$\phi^e_B = \frac{12I}{A^2} = \frac{12 \times 462198965.178}{15393.8^2} = 23.4$

The effective stiffness of simply supported beam is 

$ S_0 =\frac{3EI}{L^3} = \frac{3 \times 2.4e9 Pa \times 8.33333333 \times 10^{-6} m^4}{L^3} = C \times 60000 N/m$

The shape factor is $23.4$, so the bending stiffness of this shape is

$ S = 23.4 \times S_0 = C \times 1404333.8 N/m$

$\textbf{Derivation of Moment of Inertia for an I-Shaped Cross-Section:}$

$\text{Given an I-beam with total height} \ h, \text{width of the flanges} \ w, \text{flange thickness} \ t, \text{and web thickness} \ 2t.$

$\text{The I-beam can be divided into three rectangles:}$

1. $\text{Top flange with dimensions} \ w \times t$
2. $\text{Bottom flange with dimensions} \ w \times t$
3. $\text{Web (middle section) with dimensions} \ 2t \times (h - 2t)$

$\text{Using the formula for the moment of inertia of a rectangle:}$ 

$I = \frac{b \times h^3}{12}$

$\text{For the top flange:} $

$I_{\text{top flange, centroid}} = \frac{w \times t^3}{12}$

$\text{Using the parallel axis theorem:} $

$I_{\text{top flange}} = I_{\text{top flange, centroid}} + w \times t \times \left(\frac{h - t}{2}\right)^2$

$\text{For the bottom flange:}$

$I_{\text{bottom flange, centroid}} = \frac{w \times t^3}{12}$

$\text{Using the parallel axis theorem:} $

$I_{\text{bottom flange}} = I_{\text{bottom flange, centroid}} + w \times t \times \left(\frac{h - t}{2}\right)^2$

$\text{For the web:}$

$I_{\text{web}} = \frac{2t \times (h - 2t)^3}{12}$

$\text{The total moment of inertia of the I-beam is:} $

$I_{\text{total}} = I_{\text{top flange}} + I_{\text{bottom flange}} + I_{\text{web}} = \frac{1}{6} h^3 t (1 + 3\frac{b}{h})$


The area of the cross section is: 

$A = 2t (b + h - 2t) = 2 \times 10mm (500mm + 100mm - 2 \times 10mm) = 11600 mm^2$

The moment of inertia is:

$I = \frac{1}{6} h^3 t (1 + 3\frac{b}{h}) =  \frac{1}{6} (100mm)^3 (10mm) (1 + 3\frac{500mm}{100mm}) = 26666666.66 mm^4 = 2.66666667 \times 10^{-5} m^4 $

The shape factor is:

$\phi^e_B = \frac{12I}{A^2} = \frac{12 \times 26666666.66}{11600^2} = 2.37$

The effective stiffness of simply supported beam is 

$ S_0 =\frac{3EI}{L^3} = \frac{3 \times 2.4e9 Pa \times 8.33333333 \times 10^{-6} m^4}{L^3} = C \times 60000 N/m$

The shape factor is $2.37$, so the bending stiffness of this shape is

$ S = 2.37 \times S_0 = C \times 142200 N/m$
